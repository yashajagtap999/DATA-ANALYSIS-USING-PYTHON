```python
# To replace the missing alue with specific value:-- d1.fillna(num)

import pandas as pd
d1=pd.DataFrame({"name":["YAsh",None,"om","avi","prath"],
                 None:[10000,80000,None,60000,30000],
                 "performance":[100,90,None,30,40],
                 "age":[24,None,30,40,38]})
print(d1)
print("filled value on place of none")
print(d1.fillna(99))
```
