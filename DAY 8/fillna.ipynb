```python
# TO fill the missing value with in specific place or column:--
import pandas as pd
d1={"name":["YAsh",None,"om","avi","prath"],
                 "salary":[100000,80000,None,60000,30000],
                 "performance":[100,90,None,30,40],
                 "age":[24,None,30,40,38]}
df=pd.DataFrame(d1)
print(df)
print("replace missing value with mean:--")

df["salary"].fillna(df["salary"].mean(),inplace=True)
print(df)

df["performance"].fillna(df["performance"].mode(),inplace=True)
print(df)

df["age"].fillna(df["age"].std(),inplace=True)
print(df)
```
