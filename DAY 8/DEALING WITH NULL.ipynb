```python
#to know null values :-- d1.isnull() and to know how many null:-- d1.isnull().sum()
import pandas as pd
d1=pd.DataFrame({"name":["YAsh",None,"om","avi","prath"],
                 "salary":[100000,80000,None,60000,30000],
                 "performance":[100,90,None,30,40],
                 "age":[24,None,30,40,38]})
print(d1)
print("null values:--\n")
print(d1.isnull())

print("sum of null values\n")
print(d1.isnull().sum())
```
