```python
# sorting :-- to arrange the data in spcific manner:-- d1.sort_values(by="column",T/F)

import pandas as pd
d1={"name":["YAsh","omkar","om","avi","prath"],
                 "salary":[100000,80000,90000,60000,30000],
                 "performance":[100,90,6,30,40],
                 "age":[24,27,30,40,38]}
df=pd.DataFrame(d1)
print(df)
print("sorted_values of salary ")
df.sort_values(by="salary",ascending=False,inplace=True)
print(df)

print("sorted values of age")
df.sort_values(by="age",ascending=True,inplace=True)
print(df)
```
