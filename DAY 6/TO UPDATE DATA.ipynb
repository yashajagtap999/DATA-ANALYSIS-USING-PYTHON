```python
# to update rows and columns:--d1.loc[row index,column name,values]
import pandas as pd
d1=pd.DataFrame({"name":["YAsh","ashish","om","avi","prath"],
                 "salary":[100000,80000,50000,60000,30000],
                 "performance":[100,90,20,30,40],
                 "age":[24,20,30,40,38]})
print(d1)

d1.loc[1,"salary"]=55550
print("updated value:--\n",d1)
```

```python
# to update more than one value
d1["salary"]=d1["salary"]*10
print(d1)
```
