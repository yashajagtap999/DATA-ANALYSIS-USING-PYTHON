```python
# To create excel file:--
import pandas as pd
d1={"Name":['ram','shyam','ghamshyam'],
    "Age":[10,20,21]
   }
df=pd.DataFrame(d1)
print(df)

df.to_excel("output.xlsx")
```

```python
# To create json file:--

import pandas as pd
d1={"Name":['ram','shyam','ghamshyam'],
    "Age":[10,20,21]
   }
df=pd.DataFrame(d1)
print(df)

df.to_json("output.json")
```
