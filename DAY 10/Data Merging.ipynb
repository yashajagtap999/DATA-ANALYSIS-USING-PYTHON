```python
# To merge the data:-- pd.merge((d1,d2),on=commoncoln,how="inner/outer/left/right)

import pandas as pd
d1=pd.DataFrame({"empl_name":["Yash","Omkar","Om","Abhi"],
                 "empl_id":[101,99,102,90]})

d2=pd.DataFrame({"empl_id":[101,99,202,400],
                 "order_id":[1100,200,990,5050]})

d4=pd.merge(d1,d2,on="empl_id",how="right")
print(d4)
```
